```python
##Final
import pandas as pd
from win32com.client import Dispatch
import datetime
from datetime import datetime
from dateutil.parser import parse
from notion.client import NotionClient
from notion.block import PageBlock
from notion.block import TextBlock
from notion.block import BulletedListBlock

Last_run = "1 January 2019" ##Enter Date for last Run
Date_Last =datetime.strptime(Last_run, '%d %B %Y')
with open('D:\My Clippings.txt', encoding='utf8') as source: #Enter File Path for Kindle Clippings
    lines = []
    for line in source:
        lines.append(line)
    
    i = 0
    df = pd.DataFrame({},columns=['Title','Date','Highlight','DateStamp'],dtype=float)
    
    
    count =int(len(lines)//5)
                            
    for i in range(0,count):
            x = 5 *i
            Title = (lines[x][:len(lines[x])-2])
                        
            y = 5 *i + 1
            date_position = lines[y].find(', ')+2
            Date = (lines[y][date_position:len(lines[y])-10])
            DateStamp =datetime.strptime(Date, '%d %B %Y')
            
            z = 5 *i+3
            Highlight = (lines[z][:len(lines[z])-2])
            
            df2 = {'Title': Title, 'Date': Date, 'Highlight': Highlight, 'DateStamp':DateStamp}
            df =df.append (df2, ignore_index=True)


df.drop_duplicates( subset = "Highlight", keep="first", inplace = True)
df = df[df.DateStamp > Date_Last]
df_book_names = df['Title'].drop_duplicates(keep="first", inplace = False)    

#Enter token_v2 and Page URL
client = NotionClient(token_v2="c7e5a43e4e7bbb55a629263fbd96251e79573b11f893cc56c77918a81cbae3f0d9ba732a4c128533df953fe5b2f9893774e99f7929ee8445932ad9f80f98dbfbfa795581b4801d21caf20e58aca2")
page = client.get_block("https://www.notion.so/Bookshelf-d201e95f8ae94841a4c2930cf5f930fe")

book_count = df_book_names.count()


for i in range(0,book_count):
    newchild = page.children.add_new(PageBlock, title=df_book_names.iloc[i])
    childpage="https://www.notion.so/{}".format(newchild.id.replace("-",""))
    current_page = client.get_block(childpage)
    
    df_quotes_book = df[df.Title == df_book_names.iloc[i]]
    quote_count = df_quotes_book.count()                      
    
    for x in range(0,quote_count[1]):
            quote = current_page.children.add_new(BulletedListBlock, title=df_quotes_book.iloc[x][2])  
            quote = current_page.children.add_new(TextBlock, title='           '+ df_quotes_book.iloc[x][1])
            quote.set('format.block_color', 'gray')
            quote = current_page.children.add_new(TextBlock, title=' ')
            
```
